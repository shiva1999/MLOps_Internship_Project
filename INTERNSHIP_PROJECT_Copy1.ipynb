```python
# Import python libraries pandas and numpy
import pandas as pd
import numpy as np
# Load dataset
dataset = pd.read_csv("/content/cc_approvals.data", header=None)
# Inspect data
dataset.head()
```

```python
# Print summary statistics

dataset_description = dataset.describe()
print(dataset_description)
print("\n")
# Print DataFrame information
dataset_info = dataset.info()
print(dataset_info)
print("\n")
# Inspect missing values in the dataset
dataset.tail(10)
```

```python
# Import numpy
import numpy as np
# Inspect missing values in the dataset
print(dataset.tail(10))
# Replace the '?'s with NaN
dataset = dataset.replace('?', np.nan)
# Inspect the missing values again
dataset.tail(10)
```

```python
# Impute the missing values with mean imputation
dataset.fillna(dataset.mean(), inplace=True)
# Count the number of NaNs in the dataset to verify
print(dataset.isnull().values.sum())
```

```python
# Iterate over each column of cc_apps

for col in dataset.columns:
# Check if the column is of object type
    if dataset[col].dtypes == 'object':
# Impute with the most frequent value
        dataset = dataset.fillna(dataset[col].value_counts().index[0])
# Count the number of NaNs in the dataset and print the counts to verify
print(dataset.isnull().values.sum())
```

```python
# Import LabelEncoder

from sklearn.preprocessing import LabelEncoder
# Instantiate LabelEncoder
le = LabelEncoder()
# Iterate over all the values of each column and extract their dtypes
for col in dataset.columns:
# Compare if the dtype is object
    if dataset[col].dtypes=='object':
# Use LabelEncoder to do the numeric transformation
        dataset[col]=le.fit_transform(dataset[col])
# As we can see all features are of numeric type now
dataset.info()
```

```python
# Import train_test_split

from sklearn.model_selection import train_test_split
# Drop the features 11 and 13 and convert the DataFrame to a NumPy array
dataset = dataset.drop([dataset.columns[11], dataset.columns[13]], axis=1)
dataset = dataset.values
# Segregate features and labels into separate variables
X,y = dataset[:,0:13] , dataset[:,13]
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
# Import MinMaxScaler

from sklearn.preprocessing import MinMaxScaler
# Instantiate MinMaxScaler and use it to rescale X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

```python
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(64, activation = 'relu', input_shape=(rescaledX_test.shape[1],)))
model.add(layers.Dense(64, activation = 'relu'))
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
from keras import optimizers
model.compile(optimizer=optimizers.RMSprop(lr=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy'])
history = model.fit(rescaledX_test, y_test,
    epochs = 200,
    batch_size = 100,
    validation_data = (rescaledX_test, y_test))
```
